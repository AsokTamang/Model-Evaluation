```python
import pandas as pd
import numpy as np
from utils import build_linear_nn_model
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
df =pd.read_csv('data/data_w3_ex2.csv',header=None,names=['f1','f2','target'])
df
```

**Converting the training datas intp numpy array**

```python
X,Y = np.array(df[['f1','f2']]) , np.array(df['target'])
print('Shape of the training data is', X.shape)
print('Shape of the target data is', Y.shape)
```

**Splitting the datas into training, cross_validation and test datas**

```python
X_train,X_,Y_train,Y_ = train_test_split(X,Y,test_size=0.4,random_state=1)
X_cv,X_test,Y_cv,Y_test = train_test_split(X_,Y_,test_size=0.5,random_state=1)
print(f'The shape of training data is: {X_train.shape}')
print(f'The shape of cross_validation data is: {X_cv.shape}')
print(f'The shape of target data is: {X_test.shape}')
```

**Now its time for standard feature scaling**

```python
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)  #for training datasets
X_cv_scaled = sc.transform(X_cv)  #for cross_validation datasets
X_test_scaled = sc.transform(X_test)  #for test datasets
```

**Now its time for model evaluation, as we know for the evaluation of logistic neural network model , we regard the fraction of missclassified prediction as the misclassified error rate, so we must store the fraction of misclassified predictions, and choose the model which produces the least misclassificatiion error rate** 

```python
models = build_linear_nn_model()
for model in models:
    model.compile(
        loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),  #here we are using from_logits=True, cause we are using the linear as activation function in the output layer of each neural network model
        optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
    )
    
```
